**Looping over data structures**

- Looping over data structures (like lists, dictionaries, etc.) will be covered later

**When to use `for` vs `while`**

- `for` is used when:
  - We want to write a loop over a **set**
  - Or when we **know how many times** we want to repeat a certain task

- `while` is used when:
  - We do **not know exactly how many times** the work will be repeated
  - But we **do know the condition** under which it should end

```python
while condition:
    statement

````python
a = 5
while a > 5:
    print("Hi")
````

```python
never do this:
a = 5
while a > 4:
    print("Hi")

**Why should you never do this?**

- The condition `a > 4` is **always True**
- The value of `a` never changes
- So the loop **never ends**

What would happen if this code ran:
- `"Hi"` would be printed **forever**
- This creates an **infinite loop**

Important rule for `while` loops:
- Something inside the loop must **change the condition**
- Otherwise, the loop will never stop

````python
a = 5
while a >= 0:
    print("Hi")
    a -= 1
````

````python
# Sometime stop condition is inside the code block
a = 5
while True:
    print("Hi")
    a -= 1
    if a < 0:
        break
````

**What does this code do?**

- The loop starts with `while True`, which means the loop is **initially infinite**
- `"Hi"` is printed in each iteration
- The value of `a` is decreased by 1 every time the loop runs
- The `if` statement checks whether `a` has become negative

**What is `break`?**

- `break` is a control statement used **inside loops**
- When Python encounters `break`, it **immediately exits the loop**
- Execution continues with the code **after the loop**

**Why is `break` needed here?**

- The loop condition (`True`) never becomes False on its own
- The stopping condition is instead defined **inside the loop**
- When `a < 0` becomes True, `break` stops the loop

**Key idea**

- `break` allows us to **terminate a loop based on a condition**
- This is useful when the stopping logic cannot be written directly in the `while` condition

**Important note about `break`**

- `break` can appear **anywhere inside a loop**
- As soon as Python encounters `break`, it **immediately exits the loop**
- No further statements inside the loop are executed

**`break` in a `for` loop**

- `break` can also be used **inside a `for` loop**
- As soon as `break` is encountered, the loop **stops immediately**
- The loop does **not** continue to the next iteration

Example:

````python
for i in range(10):
    if i == 4:
        break
    print(i)
````

````python
for i in range(10):
    print(i)
    if i == 4:
        break
````

````python
for i in range(10):
    if i == 4:
        continue
    print(i)
````

````python
for i in range(10):
    if i == 4:
        pass
    print(i)
````

## 12. Dictionary

**Why lists are not good for tagged data**

- Lists are good for **ordered data**
- But lists are **not suitable for tagged or labeled data**

Example of tagged data:
- city → id, name, population, rainfall, average age

Problem with lists:
- Values are stored only by **position**
- It is hard to remember what each position means
- Data is **not easily trackable or readable**


**What is a dictionary?**

- A dictionary stores data as **key–value pairs**
- Each value is accessed using a **key**, not an index
- Dictionaries are designed for **labeled (tagged) data**

- Dictionary keys must be **unique**
  - A key **cannot be repeated**
- Keys are used to **identify values**

Key types:
- Keys should be of type:
  - `str`
  - `int`

Important notes:
- If a key is repeated, the **last value overwrites the previous one**
- Keys are used to **access values**, not positions

- `return` sends a value back to the place where the function was called
- When Python reaches `return`, the function **stops immediately**

````python
import pandas as pd

df = pd.DataFrame({
    "name": ["Ehsan", "Sara", "Jim", "Mina", "Arman", "Leila", "Neda", "Reza", "Maryam", "Omid"],
    "age": [23, 29, 31, 26, 34, 28, 41, 25, 30, 37],
    "city": ["Tehran", "London", "Tokyo", "Berlin", "Toronto", "Paris", "New York", "Istanbul", "Dubai", "Sydney"],
    "income": [31, 51, 56, 44, 72, 63, 95, 38, 88, 77],
    "gender": ["M", "F", "M", "F", "M", "F", "F", "M", "F", "M"],
    "education": ["PhD", "MA", "BA", "MA", "PhD", "BA", "MA", "BA", "MA", "PhD"],
    "department": ["Research", "Finance", "Engineering", "Marketing", "Research", "Engineering", "HR", "Finance", "Marketing", "Engineering"],
    "years_exp": [2, 6, 7, 4, 10, 5, 12, 3, 8, 11],
    "remote": [True, False, True, False, True, False, True, False, True, False],
    "performance": [4.2, 3.8, 4.5, 3.6, 4.7, 4.0, 4.1, 3.5, 4.3, 4.6],
    "start_date": ["2023-09-01", "2022-01-15", "2021-06-10", "2023-02-20", "2019-11-05", "2020-08-18", "2018-03-12", "2024-05-01", "2021-10-25", "2017-07-07"],
    "active": [True, True, True, True, True, True, False, True, True, False]
})

df["start_date"] = pd.to_datetime(df["start_date"])

df
````

````python
df["age"]
````

````python
df[["name", "city", "income"]]
````

````python
df[df["income"] >= 70]
````

````python
df[(df["city"].isin(["Tehran", "London", "Tokyo"])) & (df["active"])]
````

````python
df.sort_values(["income", "performance"], ascending=[False, False])
````

````python
df.assign(income_k=df["income"] * 1000, senior=df["years_exp"] >= 8)
````

````python
df.replace({"department": {"Engineering": "Eng", "Human Resources": "HR"}})
````

````python
df.groupby("department", as_index=False)["income"].mean()
````

````python
df.groupby("department")["income"].mean()
````

````python
df.groupby("department", as_index=False).agg(
    n=("name", "count"),
    avg_income=("income", "mean"),
    min_income=("income", "min"),
    max_income=("income", "max"),
    avg_perf=("performance", "mean"),
    remote_share=("remote", "mean")
).sort_values(["avg_income", "avg_perf"], ascending=[False, False])
````

````python
df.groupby(["department", "remote"], as_index=False).agg(
    n=("name", "count"),
    avg_income=("income", "mean"),
    avg_perf=("performance", "mean")
).sort_values(["department", "remote"])
````
